has its downsides: the scaling problem of the matrix inverse. Now, let us look at a numerical solution to the problem of finding the value of $\bw$  (aka $\btheta$) that minimizes the objective function $J$.

Again, ideally we just want to find places, where the (multi-dimensionally) gradient of $J$ is zero (here using a constant factor $\frac{2}{m}$)

$$
\ar{rl}{
  \nabla_\bw J(\bw) &= \frac{2}{m} \bX^\top \left( \bX \bw - \by \right)\\
}
$$

and numerically we calculate $\nabla_{\bw} J$ for a point in $\bw$-space, and then move along in the opposite direction of this gradient, taking a step of size $\eta$

$$ 
  \bw^{(step~N+1)} = \bw^{(step~N)} - \eta \nabla_{\bw} J(\bw)
$$

That's it, pretty simple, right (apart from numerical stability, problem with convergence and regularization, that we will discuss later).

So, we begin with some initial $\bw$, and iterate via the equation above, towards places, where $J$ is smaller, and this can be illustrated as

If we hit the/a global minimum or just a local minimum (or in extremely rare cases a local saddle point) is another question when not using a simple linear regression model: for non-linear models we will in general not see a nice convex $J$-$\bw$ surface, as in the figure above.

### Qa Solution - The Gradient Descent Method (GD)

The GD algorithm is a method for using the mean sqaured error to calculate a vector with the lowest value of the gradient in the cost function. This is done by calculating the gradient and reversing the value for the gradient, so goes towards a lower value and then recalculating the gradient based on previous vactor making it better with every iteration. We then multiply a **eta** value(Learning rate) by making it take small steps(Low Learning Rate) or take larger steps(High Learning rate).

**Eta** is used to hold the learning rate for the Algorithm, it gives the models time to fit.
In the first plot shows a low 0.02 eta, this learning rate is too low and the model goes towards the correct result but will take more iterations and in the specified method it doesn't reach a good theta with the first 10 iterations.

The second plot shows the method using a 0.1 **eta**, the result is pretty good as the method can give a good theta prediction within the first 10 steps. 

The Third plot shows the method using an 0.5 **eta**, the result is bad, because the predicted theta jumps around and overshutes the correct theta values.

### Qd Mini-batch Gradient Descent Method 

Finally explain what a __mini-batch__ SG method is, and how it differs from the two others.

Again, take a peek into the demo code below, to extract the algorithm details...and explain the __main differences__, compared with the GD and SGD.

````python
# TODO: Qd...run this code

# NOTE: code from [GITHOML], 04_training_linear_models.ipynb

theta_path_mgd = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)

print(f'mini-batch theta={theta.ravel()}')
print('OK')
````

### Qd Solution - Mini-batch Gradient Descent Method 
This method is using both GD the part about having multiple sample points to calucalting a gradient, which results in a better avarage result. and the SGD which uses multiple single random sample points in calculating the gradient decent returning a more random theta through the iterations.

With the function **np.random.permutation(m)** we shuffle our data set around and pick out a smaller batch of sample points to claculate a gradient. This is a mixed solution of doing the GD and SGD from the previous exercises. Making sure our theta become more flexible, but still reflect average of the training set.

### Qe Choosing a Gradient Descent Method

**Gradiant Decent**: utilizes all the traning data to calculate the best theta, which is very precise and if the traning data has a high correlation, then this might return a better resulat than the other gradients. However there is a large penalty for using this gradient for large data sets as we iterate through all the sample points.

**Stochatic Gradiant Decent**: Is a better method to use on larger data sets as it takes out random single samples, instead of averaging the whole data set with(GD). We then have a method of finding a good theta at low time cost. However we could be unlucky and choose a seed which picks out samples whom returns a different gradiant.

**Mini Batch gradient descent**: this technique helps the gradient to be based on all the data, through randomly packed batches of sample points. Here we reduce some of the traning time and reduce the chance of our gradient to take unwanted charatistics.
